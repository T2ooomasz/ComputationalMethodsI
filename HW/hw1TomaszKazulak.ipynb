## Test 2
Now that basic plotting works, we can do something a bit more interesting. Namely recreating something similar to the logo of the lecture.

```python
import numpy as np
import matplotlib.pyplot as plt

def count_iteration(c):
    z = 0 + 0j
    for i in range(100):
        z = z**2 + c
        if abs(z) > 2:
            break
    return i

def mandelbrot(x1,x2,y1,y2, density):
    x = np.linspace(x1, x2, density)
    y = np.linspace(y1, y2, density)
    X, Y = np.meshgrid(x, y)
    C = X + Y*1j
    Z = np.empty_like(C, dtype=int)
    for i in range(density):
        for j in range(density):
            Z[i,j] = count_iteration(C[i,j])
    plt.imshow(np.abs(Z), interpolation='nearest', cmap='hot',extent=[x1,x2,y1,y2])

mandelbrot(-2, 0.5, -1.5, 1.5, 2_000)
```
